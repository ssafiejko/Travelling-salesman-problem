## Script for generating a plot comparing the PT_SA algorithm with the SA algorithm

```python
# from Core.Simulated_Annealing import simulated_annealing
from TP_SA.src.Code.main import run_algorithm
import math
import re
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from Core.ATSP_file_reader import read_atsp_file
from Core.Simulated_Annealing import simulated_annealing

random.seed(2137)
```

```python
best_known_solution = {
    "br17": 39,
    "ft53": 6905,
    "ft70": 38673,
    "ftv33": 1286,
    "ftv35": 1473,
    "ftv38": 1530,
    "ftv44": 1613,
    "ftv47": 1776,
    "ftv55": 1608,
    "ftv64": 1839,
    "ftv70": 1950,
    "ftv90": 1579,
    "ftv100": 1788,
    "ftv110": 1958,
    "ftv120": 2166,
    "ftv130": 2307,
    "ftv140": 2420,
    "ftv150": 2611,
    "ftv160": 2683,
    "ftv170": 2755,
    "kro124p": 36230,
    "p43": 5620,
    "rbg323": 1326,
    "rbg358": 1163,
    "rbg403": 2465,
    "rbg443": 2720,
    "ry48p": 14422,
}
```

```python
def linear_decrease(Top_T, T, iteration, cool_parameter):
        return T * cool_parameter
def gaussian_probability(delta_e, temperature, prob_parameter): 
        return math.exp(-(delta_e ** 2) / (2 * (prob_parameter ** 2) * temperature))
def run_algorithm2(file_name: str):
    pattern = r"\d+"
    match = re.search(pattern, file_name)
    cool_parameter = 0.99999
    if match:
        number = match.group()
        if int(number) < 100:
            cool_parameter = 0.999991
        
    file = file_name + '.atsp'
    global_min_path, global_min, time = simulated_annealing(
        30000,linear_decrease,cool_parameter, gaussian_probability,3,0.1,file
    )
    
    return global_min
```

```python
run_algorithm2('kro124p')
```

```python
df = pd.DataFrame(columns=['file', 'algorithm', 'res'])
for file in ['ftv35', 'ftv70', 'kro124p', 'rbg403', 'br17', 'ry48p', 'rbg323', 'rbg443']:
    for _ in range(4):
        new_row = {'file': file, 'algorithm': 'SA', 'res': run_algorithm2(file)}
        df = df.append(new_row, ignore_index=True)
        _, res2 = run_algorithm(file)
        new_row = {'file': file, 'algorithm': 'TP_SA', 'res': res2}
        df = df.append(new_row, ignore_index=True)
df.to_csv('comparison.csv', index=False)
```

```python
df['file_number'] = df['file'].apply(lambda x: int(re.search(r'\d+', x).group()))

df = df.sort_values('file_number')

df['percent_above_best'] = df.apply(lambda row: (row['res'] - best_known_solution[row['file']]) / best_known_solution[row['file']] * 100, axis=1)

custom_palette = ['#7b03fc', '#0cc227']

algorithm_order = df['algorithm'].unique()

plt.figure(figsize=(12, 6), facecolor='none')  
sns.set_context("paper", font_scale=1.2)
sns.stripplot(x='file', y='percent_above_best', hue='algorithm', data=df, jitter=True, palette=custom_palette,
              order=df['file'].unique(), hue_order=algorithm_order, size=8, alpha=0.7)  
plt.xlabel('Nazwa pliku')
plt.ylabel('Procent powyej najlepszego znanego rozwizania')


plt.xticks(rotation=45)

file_labels = df['file'].unique()
for i in range(1, len(file_labels)):
    plt.axvline(x=i - 0.5, color='gray', linestyle='--', alpha=0.3)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.gca().set_facecolor('none')

plt.show()
```
